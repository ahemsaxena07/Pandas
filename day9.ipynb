```python
import yfinance as yf
import pandas as pd
import os

# Create examples folder
os.makedirs("examples", exist_ok=True)

# Download data
price = yf.download(
    ["AAPL", "MSFT", "GOOGL", "AMZN", "TSLA"],
    start="2015-01-01",
    end="2023-01-01"
)["Close"]

# Save as pickle
price.to_pickle("examples/yahoo_price.pkl")

print("✅ File saved successfully!")
print("Path:", os.path.abspath("examples/yahoo_price.pkl"))
```

```python
price = pd.read_pickle("examples/yahoo_price.pkl")
print(price.head())
```

```python
returns = price.pct_change().dropna()
```

```python
returns["MSFT"].corr(returns["TSLA"])
```
