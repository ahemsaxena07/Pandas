```python
import numpy as np
```

```python
import pandas as pd
```

```python
frame = pd.DataFrame(np.random.standard_normal((4,3)), columns = list("bde"), index = ["utah", "ohio", "texas", "oregon"])
```

```python
frame
```

```python
np.abs(frame)
```

```python
def f1(x):
    return x.max() - x.min()
```

```python
frame.apply(f1)
```

```python
frame.apply(f1, axis = "columns")
```

```python
def f2(x):
    return pd.Series([x.min(), x.max()], index = ["min", "max"])
```

```python
frame.apply(f2)
```

```python
def my_format(x):
    return f"{x:.2f}"
```

```python
frame.map(my_format)
```

```python
frame["e"].map(my_format)
```
